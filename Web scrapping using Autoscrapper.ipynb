```python
from autoscraper import AutoScraper
import pandas as pd
```

```python
url = "https://books.toscrape.com/catalogue/category/books/travel_2/index.html"
req = ["It's Only the Himalayas","£45.17"]
scraper = AutoScraper()
res = scraper.build(url,req)
sen = list()
mon = list()
for i in res:
    if i[0] == "£":
        mon.append(i)
    else:
        sen.append(i)
new = list()
for i in range(11):
    new.append(sen[i])
```

```python
data = {"title":new,"Price":mon}  #uses to convert list into data frames.
df = pd.DataFrame(data)
df
```

```python
df.to_csv("Webscrapping_the_Online_Website.csv")
```

```python
from autoscraper import AutoScraper
import pandas as pd
```

```python
url = "https://www.flipkart.com/search?q=iphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
req_list = ["APPLE iPhone 11 (White, 128 GB)","₹45,999","4.61,72,670 Ratings & 10,236 Reviews"]

flip_scrap = AutoScraper()
data = flip_scrap.build(url,req_list)
print(data)
```

```python
name = list()
price = list()
rating = list()
for i in data:
    if i[0] == "A":
        name.append(i)
    elif i[0] == "₹":
        price.append(i)
    else:
        rating.append(i)
      
```

```python
print(len(name))
print(len(price))
print(len(rating))
```
